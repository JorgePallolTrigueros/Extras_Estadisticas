```python
import requests
import time
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pymysql

try:
    connection = pymysql.connect(host='localhost',user='root',password='Kojinanjo1@',db='pelis')
    print("Conectado correctamente")
except pymysql.Error as e:
    print("No puede conectar con MySQL %d: %s" %(e.args[0], e.args[1]))
cursor = connection.cursor()

    
    
url = "https://www.play-asia.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
#driver = webdriver.Chrome(executable_path=r"C:\chromedriver.exe")
s = Service('C:/chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get(url)




time.sleep(2)

titulolista = list()
plataformaslista = list()
generoslista = list()
descripcionlista = list()
imagenlista = list()

juegosPrueba = ["Luigi's Mansion 3",'Disney Epic Mickey: The Power of Illusion',
                'Animal Crossing: New Leaf'];
los_rechazados = list()
df = pd.DataFrame()

for juego in juegosPrueba:

    driver.get(url)
    time.sleep(5)
    element=driver.find_element(By.ID,"search_input")
    element.send_keys(juego + " game")
    element.send_keys(Keys.ENTER)  
    time.sleep(5)
    response = requests.get(driver.current_url)
    time.sleep(5)
    page_source = driver.page_source

    # Verifica si la solicitud fue exitosa
    if response.status_code == 200:
        # Parsea el contenido HTML de la página web
        soup = BeautifulSoup(page_source, 'html.parser')

        # Entramos en la página del primer juego
        enlaceJuego = soup.find("div", id="n_pf_holder").find("a")['href']
        print(enlaceJuego)
        print("Navegamos a la URL del primer juego: ")
        driver.get("https://www.play-asia.com/" + enlaceJuego)
        time.sleep(5)
        
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        
        # Busqueda de nombre
        nombre = ""
        try:
            nombre = soup.find("h1", class_="p_name").text.strip()
            print("Nombre:", nombre)
            titulolista.append(nombre)
        except:
            print("No se ha encontrado el nombre")
            los_rechazados.append(juego)
            continue



        # Imagen Capturada
        main_img = soup.find("div", id="main-img")
        img_juego = main_img.find('img')
        img_url = img_juego['src']
        imagenlista.append(img_url)
            
        # Busqueda de Plataform
        plataforma = ""
        try:
            plataforma  = soup.find("div", id="pname_seg").find("a").text.strip()
        except:
            print("No se ha encontrado plataforma ")

        print("PLATAFORMA")
        print(plataforma)
        plataformaslista.append(plataforma)
            
        # Busqueda de Descripcion
        description = ""
        try:
            description  = soup.find("span", itemprop="description").text.strip()
            print("description :", description )
        except:
            print("No se ha encontrado description ") 

        descripcionlista.append(description)
            
        # Busqueda de Genero
        furtInfo = soup.find("div", id="further_info")
        scripElem = furtInfo.find_next_sibling()
        urlPattern = re.compile(r"\$.get\('([^']+)")
        match = urlPattern.search(scripElem.text)
        url_further = match.group(1)
        response = requests.get(url_further)
        time.sleep(3)
        soup = BeautifulSoup(response.text, 'html.parser')
        genero = ""
        try:
            genre_label = soup.find("i", string="Genre")
            genero = genre_label.find_next("td").text.strip()
            print("genero :", genero )
        except:
            print("No se ha encontrado genero ")             

        generoslista.append(genero)


    else:
        print("La solicitud no fue exitosa. Código de estado:", response.status_code)




    

print("Juegos Excluidos")
print(los_rechazados)
    
df["titulo"] = titulolista
df["plataformas"] = plataformaslista
df["generos"] = generoslista
df["descripcion"] = descripcionlista
df["imagen"] = imagenlista

sql = """INSERT INTO `pelis`.`emulador` (`titulo`,`plataformas`,`generos`,`descripcion`,`imagen`) VALUES (%s, %s, %s, %s, %s)"""

for ind in df.index:
    cursor.execute(sql, (df['titulo'][ind], df['plataformas'][ind], df['generos'][ind], df['descripcion'][ind], df['imagen'][ind] ))
    connection.commit()    
    
```
