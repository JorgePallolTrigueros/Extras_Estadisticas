```python
##GACER LA LISTA DE COMICS



import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep
import pymysql
import time 

try:
    connection = pymysql.connect(host='localhost',user='root',password='Kojinanjo1@',db='pelis')
    print("Conectado correctamente")
except pymysql.Error as e:
    print("No puede conectar con MySQL %d: %s" %(e.args[0], e.args[1]))

cursor = connection.cursor()

print("CONTINUEEEEEEEEEEEEEEEE")

df = pd.DataFrame()


Nombres = list()
Imagen = list()
Autors = list()
descripciones = list()
cronologias = list()
generos = list()


#response = requests.get("https://tebeosfera.com/", verify=False)
#soup = BeautifulSoup(response.text, "html.parser")

#driver = webdriver.Chrome(executable_path=r"C:\chromedriver.exe")
s = Service('C:/chromedriver.exe')
driver = webdriver.Chrome(service=s)
url = "https://tebeosfera.com/"
driver.get(url)
# Input Text cada uno de los elementos de la lista

nombrelista = list()
imagenlista = list()
descripcionlista = list()
autorlista = list()

# Practicamos sobre una lista de Comics

#comicsPrueba = list()
#comicsPrueba.append(Comiccol3)
#comicsPrueba.append('ALIX SENATOR')
#comicsPrueba.append('Alpha flight')

comicsPrueba=['Arion, Lord of Atlantis - 18 - Blood Brothers', 'Comics', 'Crónicas Bárbaras', 'El bosque de las virgenes ', 
              'El Gran Bush  ', 'Héroes del Humor 1 retapado  ', 'Héroes del Humor 3 retapado  Incompleto', 
              'jiraishin diablo 3   2', 'MEDICI - 3 - Julio', 'Revista Petete ', 'Selecciones Vértice 04 grapa ', 'VOLÁTIL',
              'WOUNDED nº 1', '5 tierras', 'A La Recherche Du Temps Perdu', 'A la sombra del convoy', 'Alix Senator', 
              'Ambulancia 13', 'ANGEL WINGS', 'aristocratas', 'Bandera negra', 'BARBANZONS', 'Bill y Bolita', 'Blacksad', 
              'Blanco y Negro ¡guay!', 'bob robert', 'Brechas', 'BRIGGS LAND', 'CAIRO', 'CARVALHO', 'CASACAS AZULES', 
              'Cavall Fort', 'Centaurus', 'Chicha Tato y Clodoveo', 'Conquistador', 'Cuadernos Ucranianos', 'DAMOCLES'];


#comicsPrueba=['El cartero de las mujeres'];


comic_no_encontrados= list()
df = pd.DataFrame()

for comic in comicsPrueba:
    
    driver.get(url)
    time.sleep(5)
    element=driver.find_element(By.ID,"busqueda")
    element.send_keys(comic)
    element.send_keys(Keys.ENTER)  
    time.sleep(5)
    
    try:
        parentElement = driver.find_element(By.CLASS_NAME, "linea_resultados")
        enlace = parentElement.find_elements(By.TAG_NAME, "a")
        
        # Si el comic No existe Mas adelante su linea en el Csv sera puesta todos sus campos como nulls

        ############################  IMAGEN #####################################################

        #Cogemos el enlace que es el elemento 1 del array  y cogesmos su foto

        foto = enlace[0].get_attribute('href')

        if "promociones/" in foto or "autores/" in foto:
            print("Comic es de tipo promociones o autores... se ignora")
        else:
            try:
                ##########################  DETALLE ##################################################
                #En el enlace 2 entramos en el detalle de el Comic para coger el resto de las listas
                print("Navegamos a la URL: ")
                print(enlace[1].get_attribute('href'))
                driver.get(enlace[1].get_attribute('href'))
                time.sleep(5)    
                print("El comic no es de tipo promociones ni autores, cogemos datos!")
                print("DEBUG: despues detalle")
                #####################################################################
                # cogemos la sinopsis (Descripcion)
                sinopsisHtml = None
                try:
                    print("DEBUG: sinopsis A")
                    sinopsis = driver.find_element(By.CLASS_NAME, "T3WISIWISI")
                    print("DEBUG: sinopsis B")
                    sinopsisHtml = sinopsis.get_attribute('innerHTML')
                    print("DEBUG: sinopsis C")
                except:
                    print("DEBUG: No hay sinopsis")
                    sinopsisHtml = None
                    print("DEBUG: sinopsis D")

                ####################################################################
                # cogemos el autor

                print("DEBUG: cogemos el autor")
                enlacesNav = driver.find_element(By.CLASS_NAME, "nav-tabs")
                print("DEBUG: cogemos el autor tabs")
                enlacesTab = enlacesNav.find_elements(By.TAG_NAME, "a")
                print("DEBUG: enlaces")
                try:
                    enlacesTab[2].click()
                except:
                    enlacesTab[1].click()
                print("DEBUG: antes sleep")
                time.sleep(15)

                autoresComic = list()
                try:
                    autoresContenedor = driver.find_element(By.ID,"tab3")
                    autores = autoresContenedor.find_elements(By.TAG_NAME, "strong")
                    for autorStrong in autores:
                        autorNombre = autorStrong.find_elements(By.TAG_NAME, "a")
                        autorStrongEl = autorNombre[0].get_attribute('innerHTML')
                        autoresComic.append(autorStrongEl)
                except:
                    print("No hay autores en tab3")
                    try:
                        autoresContenedor = driver.find_element(By.ID,"tab2")
                        autores = autoresContenedor.find_elements(By.TAG_NAME, "strong")
                        for autorStrong in autores:
                            autorNombre = autorStrong.find_elements(By.TAG_NAME, "a")
                            autorStrongEl = autorNombre[0].get_attribute('innerHTML')
                            autoresComic.append(autorStrongEl)
                    except:
                        print("No hay autores en tab2")
                print("se va a hacer append a autorlista")
                print(autoresComic)
                
                nombrelista.append(comic)
                imagenlista.append(foto)
                descripcionlista.append(sinopsisHtml)
                autorlista.append(','.join(autoresComic));
                
            except:
                print("No hay informacion suficiente del comic...")
                print("DEBUG: COMIC SIN INFORMACION SUFICIENTE!!!!!!!!")
                comic_no_encontrados.append(comic)
        
    except:
        print("DEBUG: COMIC NO ENCONTRADO!!!!!!!!")
        comic_no_encontrados.append(comic)

            
####################################################################       
    
print("Datos Recogidos antes de meter en BBDD")
print(nombrelista)
print(imagenlista)
print(descripcionlista) # Coge el del elemento 1
print(autorlista) # Coge la lista de autores

print(len(nombrelista))
print(len(imagenlista))
print(len(descripcionlista))
print(len(autorlista))

df["nombre"] = nombrelista
df["imagen"] = imagenlista
df["descripcion"] = descripcionlista
df["autor"] = autorlista

    
sql = """INSERT INTO `pelis`.`comic` (`nombre`,`imagen`,`descripcion`,`autor`) VALUES (%s, %s, %s, %s)"""

for ind in df.index:
    cursor.execute(sql, (df['nombre'][ind], df['imagen'][ind], df['descripcion'][ind], df['autor'][ind] ))
    connection.commit() 

df.head(2)
####################################################################  
# ver la lista de los no encontrados

print(comic_no_encontrados)
```
