```python
manga=['Black paradox', 'Blue Spring - Tomo Único', 'Dark', 'Deatte 5 Byou de Battle - 1-15 - Absorbiendo Mangas', 'El jurado 13 - Milky y Nakatsuka '];
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time 
import pymysql
try:
    connection = pymysql.connect(host='localhost',user='root',password='Kojinanjo1@',db='pelis')
    print("Conectado correctamente")
except pymysql.Error as e:
    print("No puede conectar con MySQL %d: %s" %(e.args[0], e.args[1]))

cursor = connection.cursor()

print("CONTINUE")

df = pd.DataFrame()

Nombres = list()
Imagen = list()
Autors = list()
descripciones = list()
cronologias = list()
generos = list()




url = "https://es.ninemanga.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
#driver = webdriver.Chrome(executable_path=r"C:\chromedriver.exe")
s = Service('C:/chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get(url)
 
    # cosas PARA RELLENAR
    
nombrelista = list()
imagenlista = list()
descripcionlista = list()
autorlista = list()

comicsPrueba = list()
       
       
#comicsPrueba.append('Saint seiya - Poseidon')
#comicsPrueba.append('Saint Seiya')
#comicsPrueba.append('Saint Seiya - Next Repeticion')

comicsPrueba =["00dfghjfdgjfdgjfd-V", "Brother X Brother"];
los_rechazados = list()


for comic in comicsPrueba:
    element = driver.find_element(By.ID, "lookupwords")
    element.clear()
    
    print(comic)
    element.send_keys(comic)
    element.send_keys(Keys.ENTER)
    time.sleep(5)
    
    #obtain window handle of browser in focus
    # p = driver.current_window_handle
    #obtain parent window handle
    parent = driver.window_handles[0]
    #obtain browser tab window
    chld = driver.window_handles[1]
    #switch to browser tab
    driver.switch_to.window(chld)

    url_actual = driver.current_url
    print("URL actual:", url_actual)
    
    try:
        parentElement = driver.find_element(By.CLASS_NAME, "direlist")
        print("SI se ha encontrado direlist")
    except:
        print("NO se ha encontrado direlist") 
    
    enlaces = parentElement.find_elements(By.CLASS_NAME, "bookname")
    enlaceFinal = None
    for enlace in enlaces:
        valorHtml = enlace.get_attribute('innerHTML')
        if (valorHtml.lower() == comic.lower()):
            enlaceFinal = enlace
        
    if enlaceFinal is None:
        # Se pone en la lista de rechazados
        print("No ha encontrado enlace final... RECHAZADO")
        los_rechazados.append(comic)
        print("volvemos a la pagina principal y buscamos el siguiente...")
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        
    else:

        print("Enlace Final:")
        print(enlaceFinal)

        enlaceFinal.click()
        time.sleep(5)


        mangaContenedor = driver.find_element(By.CLASS_NAME,"bookintro")
        imagen = mangaContenedor.find_element(By.TAG_NAME, "img")
        foto = imagen.get_attribute("src")


        mangaContenedor = driver.find_element(By.CLASS_NAME,"bookintro")

        descripcion = ""
        valorHtml_descripcion = ""
        try:
            descripcion = mangaContenedor.find_element(By.TAG_NAME, "p")
            valorHtml_descripcion = descripcion.get_attribute('innerHTML')
        except:
              print("No se encuentra descripcion...")



        ########################################

        mangaContenedor = driver.find_element(By.CLASS_NAME,"bookintro")
        enlas = mangaContenedor.find_elements(By.TAG_NAME, "a")
        nombreAutor = ""
        for enla in enlas:
            try:
                attrEnla = enla.get_attribute('itemprop')
                if (attrEnla == 'author'):
                    nombreAutor = enla.get_attribute('innerHTML')

            except:
                  print("En este enlace no hay atributo itemprop...")
        
        nombrelista.append(comic)
        imagenlista.append(foto)
        descripcionlista.append(valorHtml_descripcion)
        autorlista.append(nombreAutor)
        
        print("volvemos a la pagina principal y buscamos el siguiente...")
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        #driver.get(url)
        time.sleep(5)

                

print("Longitud de nombrelista:", len(nombrelista))
print("Longitud de imagenlista:", len(imagenlista))
print("Longitud de descripcionlista:", len(descripcionlista))
print("Longitud de autorlista:", len(autorlista))


df["nombre"] = nombrelista
df["imagen"] = imagenlista
df["descripcion"] = descripcionlista
df["autor"] = autorlista
    
sql = """INSERT INTO `pelis`.`manga` (`nombre`,`imagen`,`descripcion`,`autor`) VALUES (%s, %s, %s, %s)"""

for ind in df.index:
    cursor.execute(sql, (df['nombre'][ind], df['imagen'][ind], df['descripcion'][ind], df['autor'][ind] ))

connection.commit() 
    

print("FIn del bucle")
df.head(2)


##########################################
#Los rechazafos
```
